```python
def get_recommendations_by_id(anime_id, cosine_sim=cosine_sim):

    idx = anime2.index[anime2['anime_id'] == anime_id].tolist()[0]#getting the index of the anime that matches anime_id

    sim_scores = list(enumerate(cosine_sim[idx]))#finding pairwise similarity

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)#sorting

    sim_scores = sim_scores[1:6]#getting top 5 animes according to similarity score.

    anime_indices = [i[0] for i in sim_scores]#finding the indices of top 5 animes.

    return anime2.loc[anime_indices, 'anime_id']
```

# Approach 2: Genre based recommendation system

```python
columns_to_keep = ['anime_id', 'genres', 'score_rank', 'popularity_rank', 'total_count']
anime3 = anime[columns_to_keep].copy()
anime3
```

```python
anime3['genres'] = anime3['genres'].str.split(',')
anime_3_exploded = anime3.explode('genres').reset_index(drop=True)

# One-hot encode the genres
genres_encoded = pd.get_dummies(anime_3_exploded['genres'], prefix='genre')
anime_3_final = pd.concat([anime_3_exploded.drop(columns=['genres']), genres_encoded], axis=1)
```

```python
anime_3_final
```

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_normalize = ['score_rank', 'popularity_rank', 'total_count']
scaler = MinMaxScaler()

# Normalizing the numerical columns:
anime_3_final[columns_to_normalize] = scaler.fit_transform(anime_3_final[columns_to_normalize])
```

```python
anime_3_final['composite_score'] = (anime_3_final['score_rank'] + anime_3_final['popularity_rank'] + anime_3_final['total_count']) / 3
```

```python
# Melting the DataFrame to work with genres as a single column again
melted_df = anime_3_final.melt(id_vars=['anime_id', 'composite_score'],
                                var_name='genre', value_name='is_genre')

# Filtering out rows where is_genre is 0
melted_df = melted_df[melted_df['is_genre'] == 1]
melted_df['genre_composite_rank'] = melted_df.groupby('genre')['composite_score'].rank(method='min', ascending=False)
melted_df.sort_values(by=['genre', 'genre_composite_rank'], inplace=True)

```

```python
melted_df
```

```python
def get_genre_based_recommendations(favorite_anime_ids, melted_df,anime2):

    # Finding genres of the user's favorite animes
    favorite_genres = melted_df[melted_df['anime_id'].isin(favorite_anime_ids)]['genre'].unique()
    recommendations = []
    for genre in favorite_genres:
        # Finding the top-ranked anime in the genre that's not already a favorite
        top_anime_in_genre = melted_df[(~melted_df['anime_id'].isin(favorite_anime_ids)) &
                                       (melted_df['genre'] == genre)].sort_values(by='genre_composite_rank').head(1)
        recommendations.append(top_anime_in_genre)

    # Combining all recommendations into a single DataFrame
    recommendations_df = pd.concat(recommendations)

    # Getting recommended anime names
    recommended_anime_ids = recommendations_df['anime_id']
    recommended_anime_names = [get_anime_name_by_id(anime2, anime_id) for anime_id in recommended_anime_ids]
    recommendations_df['anime_name'] = recommended_anime_names


    for idx, row in recommendations_df.iterrows():
        print(f"Genre: {row['genre']}")
        print(f"Top Recommendation: {row['anime_name']} (ID: {row['anime_id']})")
        print()

    return recommendations_df
```

```python
import numpy as np
def recommend_similar_anime(input_anime_title,anime4, top_n=10):
    # Finding the index of the input anime
    input_anime_index = anime4.index[anime4['title'] == input_anime_title].tolist()

    if not input_anime_index:
        print("Input anime '{}' not found.".format(input_anime_title))
        return []

    # Getting Word2Vec vector representation of the input anime
    input_anime_vector = anime4.loc[input_anime_index[0], 'word2vec_vectors']

    # Converting the list of Word2Vec vectors into a 2D NumPy array
    anime_vectors = np.array([vec if vec is not None else np.zeros_like(input_anime_vector) for vec in anime4['word2vec_vectors']])

    # Excluding input anime from anime vectors
    anime_vectors = np.delete(anime_vectors, input_anime_index, axis=0)

    # Calculating cosine similarity between input anime and all other anime
    similarity_scores = cosine_similarity([input_anime_vector], anime_vectors)[0]

    # Sorting anime by similarity scores
    similar_anime_indices = similarity_scores.argsort()[::-1]

    # Selecting top N similar anime
    top_similar_anime_indices = similar_anime_indices[:top_n]

    # Getting top N similar anime titles
    top_similar_anime_titles = anime4.iloc[top_similar_anime_indices]['title'].tolist()

    return top_similar_anime_titles

'''input_anime_title = "Berserk: Ougon Jidai-hen I - Haou no Tamago"
top_similar_anime = recommend_similar_anime(input_anime_title, anime4)

if top_similar_anime:
    print("Top 10 recommended anime similar to '{}' are:".format(input_anime_title))
    for anime_title in top_similar_anime:
        print(anime_title)'''
```

```python
import numpy as np
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models import KeyedVectors

# Converting GloVe format to Word2Vec format
glove_input_file = '/content/glove/glove.6B.100d.txt'
word2vec_output_file = 'glove.6B.100d.word2vec'
glove2word2vec(glove_input_file, word2vec_output_file)

# Loading the GloVe embeddings
try:
    glove_model = KeyedVectors.load_word2vec_format(word2vec_output_file, binary=False)
except FileNotFoundError:
    print("GloVe file not found. Please check the file path.")
    exit()
except Exception as e:
    print("Error loading GloVe embeddings:", e)
    exit()

def get_word_vector(word, model, embedding_dim):
    """Retrieve the word vector from the model or assign a random vector if word is out-of-vocabulary."""
    try:
        return model[word]
    except KeyError:
        # Handling out-of-vocabulary words
        return np.random.randn(embedding_dim)

def vectorize_synopsis(synopsis, model, embedding_dim):
    """Vectorize the synopsis using GloVe embeddings."""
    if isinstance(synopsis, str):  # Checking if synopsis is a string
        tokenized_synopsis = synopsis.split()  # Assuming synopsis is already preprocessed and tokenized
        word_vectors = [get_word_vector(word, model, embedding_dim) for word in tokenized_synopsis]
        if word_vectors:
            return np.mean(word_vectors, axis=0)  # Average the word vectors to get the synopsis vector
        else:
            return np.zeros(embedding_dim)  # Returning zero vector for empty synopsis
    else:
        return np.zeros(embedding_dim)  # Returning zero vector for NaN synopsis


# Converting each synopsis into a fixed-length vector representation
synopsis_vectors = []
embedding_dim = len(glove_model['word'])  # Getting the dimensionality of the embeddings
for synopsis in anime2['synopsis']:
    synopsis_vector = vectorize_synopsis(synopsis, glove_model, embedding_dim)
    synopsis_vectors.append(synopsis_vector)

# Converting to numpy array
synopsis_vectors = np.array(synopsis_vectors)

# Displaying the GloVe representation of the synopsis text
print(synopsis_vectors)

```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(target_synopsis_vector, synopsis_vectors, anime_data, target_anime_index, top_n=10):
    cosine_similarities = cosine_similarity([target_synopsis_vector], synopsis_vectors).flatten()

    similar_anime_indices = cosine_similarities.argsort()[::-1]

    similar_anime_indices = [index for index in similar_anime_indices if index != target_anime_index]

    # Getting the genres of the target anime
    target_genres = anime_data.loc[target_anime_index, 'genres']

    recommended_anime_indices = [index for index in similar_anime_indices if any(genre in target_genres for genre in anime_data.loc[index, 'genres'])]

    recommended_anime_titles = anime2.iloc[recommended_anime_indices]['title'].values.tolist()

    # Getting cosine similarity scores for recommended anime
    cosine_similarities_recommended = cosine_similarities[recommended_anime_indices]

    # Getting genres of recommended anime
    recommended_anime_genres = anime2.iloc[recommended_anime_indices]['genres'].values.tolist()

    # Combining titles, genres, and cosine similarity scores
    recommended_anime_info = [(title, genres, cosine) for title, genres, cosine in zip(recommended_anime_titles, recommended_anime_genres, cosine_similarities_recommended)]

    return recommended_anime_info[:top_n]

target_anime_title = 'One Piece'  # Specify the target anime title
target_anime_index = anime2[anime2['title'] == target_anime_title].index[0]  # Get the index of the target anime
target_synopsis_vector = synopsis_vectors[target_anime_index]  # Get the synopsis vector of the target anime

# Recommending anime similar to the target anime
recommended_anime = recommend_anime(target_synopsis_vector, synopsis_vectors, anime2, target_anime_index)

print("Recommended anime for", target_anime_title, ":")
for title, genres, cosine_similarity in recommended_anime:
    print("Title:", title)
    print("Genres:", genres)
    print("Cosine Similarity:", cosine_similarity)
    print()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def intra_list_diversity(Y):
    disimilarity_sum = 0
    num_pairs = 0

    for i in range(len(Y)):
        for j in range(i+1, len(Y)):
            item1 = Y[i].reshape(1, -1)
            item2 = Y[j].reshape(1, -1)
            disimilarity = 1 - cosine_similarity(item1, item2)
            disimilarity_sum += disimilarity
            num_pairs += 1

    if num_pairs > 0:
        avg_dissimilarity = disimilarity_sum / num_pairs
    else:
        avg_dissimilarity = 0

    return avg_dissimilarity

# Example :
list2 = recommend_similar_anime('One Piece', anime4, top_n=10)
item_vectors = vectorizer.fit_transform(list2)
item_vectors_dense = item_vectors.toarray()


```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_similar_anime(input_anime_title, anime4, top_n=10):
    # Finding the index of the input anime
    input_anime_index = anime4.index[anime4['title'] == input_anime_title].tolist()

    # Checking if input anime exists in the DataFrame
    if not input_anime_index:
        print("Input anime '{}' not found.".format(input_anime_title))
        return []

    # Getting Word2Vec vector representation of the input anime
    input_anime_vector = anime4.loc[input_anime_index[0], 'word2vec_vectors']

    # Converting the list of Word2Vec vectors into a 2D NumPy array
    anime_vectors = np.array([vec if vec is not None else np.zeros_like(input_anime_vector) for vec in anime4['word2vec_vectors']])

    # Excluding input anime from anime vectors
    anime_vectors = np.delete(anime_vectors, input_anime_index, axis=0)

    # Calculating cosine similarity between input anime and all other anime
    similarity_scores = cosine_similarity([input_anime_vector], anime_vectors)

    # Flattening the similarity scores array
    similarity_scores = similarity_scores.flatten()

    similar_anime_indices = similarity_scores.argsort()[::-1]
    top_similar_anime_indices = similar_anime_indices[:top_n]
    top_similar_anime_titles = anime4.iloc[top_similar_anime_indices]['title'].tolist()

    return top_similar_anime_titles

# Example:
list2 = recommend_similar_anime('One Piece', anime4, top_n=10)
print(list2)
```
